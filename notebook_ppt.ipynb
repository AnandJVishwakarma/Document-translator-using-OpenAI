```python
def translate_text(text, target_language="French"):
    if not text.strip():
        return text  # Skip empty strings

    # OpenAI translation prompt
    prompt = f"""
    You are a professional translator. Translate the following text into {target_language}, maintaining formatting and tone.
    
    Text: "{text}"
    Translation:
    """
    from openai import Client

    try:
        client = Client()
        response=client.chat.completions.create(
            model="gpt-4",  # Use "gpt-4" for better translations; "gpt-3.5-turbo" is also an option.
            messages=[
                {"role": "system", "content": "You are an expert translator."},
                {"role": "user", "content": prompt},
            ],
            max_tokens=500,
            temperature=0.7,
        )
        print("Coversions: ", response.choices[0].message.content)
        # response = openai.ChatCompletion.create(
        #     model="gpt-4",  # Use "gpt-4" for better translations; "gpt-3.5-turbo" is also an option.
        #     messages=[
        #         {"role": "system", "content": "You are an expert translator."},
        #         {"role": "user", "content": prompt},
        #     ],
        #     max_tokens=500,
        #     temperature=0.7,
        # )
        # Extract the translated text from the response
        translated_text = response.choices[0].message.content.strip()
        return translated_text
    except Exception as e:
        print(f"Error translating text: {e}")
        return text  # Return original text in case of error
```

```python
# # Function to process PowerPoint and translate content
# def translate_presentation(input_file, output_file, target_language="French"):
#     # Load the PowerPoint presentation
#     presentation = Presentation(input_file)

#     # Loop through slides and their elements
#     for slide in presentation.slides:
#         for shape in slide.shapes:
#             # Check if the shape contains text
#             if shape.has_text_frame:
#                 for paragraph in shape.text_frame.paragraphs:
#                     # Translate the paragraph text
#                     original_text = paragraph.text
#                     translated_text = translate_text(original_text, target_language)
#                     print("translated_text: ", translated_text)
#                     paragraph.text = translated_text  # Replace with translated text

#     # Save the translated presentation
#     presentation.save(output_file)
#     print(f"Translated presentation saved to {output_file}")


def translate_presentation(input_file, output_file, target_language="French"):
    # Load the PowerPoint presentation
    presentation = Presentation(input_file)

    # Loop through slides and their elements
    for slide in presentation.slides:
        for shape in slide.shapes:
            # Check if the shape contains text
            if shape.has_text_frame:
                for paragraph in shape.text_frame.paragraphs:
                    original_text = paragraph.text
                    if original_text.strip():  # Process non-empty paragraphs
                        print(f"Original: {original_text}")  # Debug: Show original text
                        
                        # Translate the paragraph text
                        translated_text = translate_text(original_text, target_language)
                        print(f"Translated: {translated_text}")  # Debug: Show translated text

                        # Replace the original text with the translated text
                        paragraph.text = translated_text
                        print(f"Updated paragraph: {paragraph.text, translated_text}")  # Debugging
    # Save the translated presentation
    presentation.save(output_file)
    print(f"Translated presentation saved to {output_file}")
```

```python

# Main function to execute the POC
if __name__ == "__main__":
    # Input and output file paths
    input_pptx = "input_presentation.pptx"  # Replace with your file path
    output_pptx = "translated_presentation.pptx"
    target_language = "Hindi"  # Specify the target language

    # Translate the presentation
    if os.path.exists(input_pptx):
        translate_presentation(input_pptx, output_pptx, target_language)
    else:
        print(f"Input file '{input_pptx}' not found.")
```
